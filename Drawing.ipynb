```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayImage(image, title):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
surface = np.zeros((300,300,3), dtype="uint8")
```

### Draw Lines

```python


red = (255, 0, 0)
cv2.line(surface, (0,0), (300,300), red, 5)
displayImage(surface, "red line")

```

```python
green = (0, 255, 0)
blue = (0, 0, 255)
cv2.line(surface, (0,300), (300,0), green, 5)
displayImage(surface, "green and red line")
```

### Draw Rectangles

```python
cv2.rectangle(surface, (10,10), (60,70), green, 3)
displayImage(surface, "with recangle")
```

### DRAW Circles

```python
cv2.circle(surface, (150,150), 40, blue, 2)
displayImage(surface, "with cicle")
```

```python
for i in range(10, 100, 10):
    cv2.circle(surface, (150,150), i, red, 2)
    
displayImage(surface, "with concentric circles")
```

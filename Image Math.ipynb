```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayImage(image, title):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread("./data/orca.jpg")
displayImage(image,"original")
```

```python
P = np.ones(image.shape, dtype = "uint8") * 100
added = cv2.add(image, P)
displayImage(added, "increased pixel intensity by 100")
 
```

```python
# darker
M = np.ones(image.shape, dtype = "uint8") * 60
subtracted = cv2.subtract(image, M)
displayImage(subtracted, "reduced pixel intensity by 60")
```

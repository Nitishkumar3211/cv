```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayImage(image, title):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread("./data/gg2.jpg")
displayImage(image,"original")
```

```python
mask = np.zeros(image.shape[:2], dtype="uint8")
cv2.rectangle(mask, (100, 300), (400, 750), 255, -1)
displayImage(mask, "before applying mask")
 
# Apply our mask -- notice how only the person in the image is cropped out
masked = cv2.bitwise_and(image, image, mask=mask)
displayImage(masked, "after apllying mask")
```

```python
mask = np.zeros(image.shape[:2], dtype="uint8")
cv2.circle(mask, (300, 520), 200, 255, -1)
displayImage(mask, "before applying mask")
 
# Apply our mask -- notice how only the person in the image is cropped out
masked = cv2.bitwise_and(image, image, mask=mask)
displayImage(masked, "after apllying mask")

```

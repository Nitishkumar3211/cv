```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayImage(image, title):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread("./data/RGB.png")
displayImage(image,"original")
```

```python
(R, G, B) = cv2.split(image)

# show each channel individually
displayImage(R,"Red")
displayImage(G, "Green")
displayImage(B, "Blue")
```

```python

# merge the image back together again
merged = cv2.merge([R, G, B])
displayImage(merged, "merged")
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayColorImage(image, title):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), interpolation="nearest")
    plt.title(title)
    plt.show()
```

```python
def displayGrayScaleImage(image, title):
    plt.imshow(image, cmap='gray', vmin = 0.0, vmax = 255.0)
    plt.title(title)
    plt.show()
```

```python
def getAutoEdge(image, sigma=0.23):
    v = np.median(image)
    l = int(max(0, (1.0 - sigma) * v))
    u = int(min(255, (1.0 + sigma) * v))
    e = cv2.Canny(image, l, u)
    return e
```

```python
# Read Image 
image = cv2.imread("./data/lena.jpg")
displayColorImage(image, "Original Image")
```

```python
# Convert it to grayscale
gray = cv2.cvtColor(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), cv2.COLOR_BGR2GRAY)
displayGrayScaleImage(gray, "Gray Scale Image")
```

```python

# Apply gaussian blur
blurred = cv2.GaussianBlur(gray, (5,5), 0)
displayGrayScaleImage(blurred, "After applying gaussian blur")
```

```python
#Specify Hyestersis threshold manually
postCanny = cv2.Canny(blurred, 50, 255)
displayGrayScaleImage(postCanny, "After applying Canny edge detector")
```

```python
autocanny = getAutoEdge(blurred)
displayGrayScaleImage(autocanny, "After applying Canny edge detector")
```

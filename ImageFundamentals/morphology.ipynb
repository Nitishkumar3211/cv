```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayColorImage(image, title):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), interpolation="nearest")
    plt.title(title)
    plt.show()
```

```python
def displayGrayScaleImage(image, title):
    plt.imshow(image, cmap='gray', vmin = 0.0, vmax = 255.0)
    plt.title(title)
    plt.show()
```

```python
# Read Image 
image = cv2.imread("./data/Amazonlogo.jpg")
displayColorImage(image, "Original Image")
  
```

```python
# Convert itto grayscale
gray = cv2.cvtColor(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), cv2.COLOR_BGR2GRAY)
displayGrayScaleImage(gray, "Gray Scale Image")
```

### Erosion Effect

```python
# apply a series of erosions
for i in range(0, 10):
    eroded = cv2.erode(gray.copy(), None, iterations=i + 1)
    displayGrayScaleImage(eroded, "Eroded {} times".format(i + 1))
```

### Dilation Effect

```python
# apply a series of dilations
for i in range(0, 10):
    dilated = cv2.dilate(gray.copy(), None, iterations=i + 1)
    displayGrayScaleImage(dilated, "dilated {} times".format(i + 1))
```

```python
# Read Image 
image = cv2.imread("./data/Amazonlogo.jpg")
displayColorImage(image, "Original Image")
     
```

```python
# Convert itto grayscale
gray = cv2.cvtColor(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), cv2.COLOR_BGR2GRAY)
displayGrayScaleImage(gray, "Gray Scale Image")

```

### Opening

```python

filterSizes = [(5, 5), (9, 9), (13, 13)]
for filterSize in filterSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, filterSize)
    opening = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel)
    #opening = cv2.morphologyEx(gray, cv2.MORPH_CROSS, kernel)
    #opening = cv2.morphologyEx(gray, cv2.MORPH_ELLIPSE, kernel)
    displayGrayScaleImage(opening, "Opening: ({}, {})".format(filterSize[0], filterSize[1]))
```

### Black HAT 
## Find dark regions on white background

```python

rectangle = cv2.getStructuringElement(cv2.MORPH_RECT, (15, 7))
bh = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, rectangle)
displayGrayScaleImage(bh, "After applying black hat")
```

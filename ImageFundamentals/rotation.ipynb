```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayImage(image, title):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
#load image from disk
image = cv2.imread("./data/person_image.jpg")
displayImage(image, "original")



```

```python
#get center of image
(height, width) = image.shape[:2]
(centerX, centerY) = (width / 2, height / 2)
 
# rotate our image by 30 degrees in counterclockwise direction, scale by 1.0 
Modified = cv2.getRotationMatrix2D((centerX-50, centerY-60), 45, 1.0)
postRotation = cv2.warpAffine(image, Modified, (width, height))
displayImage(postRotation, "Rotated by 30 Degrees")
 
```

```python

# rotate our image by 90 degrees in clockwise direction
Modified = cv2.getRotationMatrix2D((centerX, centerY), -90, 1.0)
postRotation = cv2.warpAffine(image, Modified, (width+250, height))
displayImage(postRotation, "Rotated by -90 Degrees")
```

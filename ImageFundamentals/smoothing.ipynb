```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayColorImage(image, title):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), interpolation="nearest")
    plt.title(title)
    plt.show()
```

```python
def displayGrayScaleImage(image, title):
    plt.imshow(image, cmap='gray', vmin = 0.0, vmax = 255.0)
    plt.title(title)
    plt.show()
```

```python
# Read Image 
image = cv2.imread("./data/tajmahal.jpg")
displayColorImage(image, "Original Image")
  
```

### Average Blurring

```python
kernelSizes = [(5,5), (9,9), (17,17)]
for (x,y) in kernelSizes:
    blurred = cv2.blur(image,(x,y))
    displayColorImage(blurred, "Average ({}, {})".format(x, y))
```

### Gaussian blur

```python
for (x,y) in kernelSizes:
    gaussblurred = cv2.GaussianBlur(image,(x,y),0)
    displayColorImage(gaussblurred, "Gaussian Blur kernel size ({}, {})".format(x, y))
```

### Median Blurring

```python
# Read Image 
image = cv2.imread("./data/peppernoise.jpg")
displayColorImage(image, "Original Image")
```

```python
squareKernels = (3,5,9)

for x in squareKernels:
    mblurred = cv2.medianBlur(image,x)
    displayColorImage(mblurred, "mblurred ({}, {})".format(x,x))
```

### Bilateral Blurring

```python
# Read Image 
image = cv2.imread("./data/tajmahal.jpg")
displayColorImage(image, "Original Image")
```

```python
p = [(11, 21, 7), (11, 41, 21), (11, 61, 39)]
 
# loop over the diameter, sigma color, and sigma space
for (d, sC, sS) in p:
    # apply bilateral filtering and display the image
    blurred = cv2.bilateralFilter(image, d, sC, sS)
    title = "Blurred d={}, sc={}, ss={}".format(d, sC, sS)
    displayColorImage(blurred, title)
```

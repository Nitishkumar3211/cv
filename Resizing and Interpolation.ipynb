```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayImage(image, title):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
#load image from disk
image = cv2.imread("./data/obama.jpg")
displayImage(image, "original")


```

```python
#Resizing Width without preserving aspect ratio
dimNewImage = (200, image.shape[1])
# perform the actual resizing of the image
resizedImage = cv2.resize(image, dimNewImage, interpolation=cv2.INTER_AREA)
displayImage(resizedImage,"Reduced width to 200")
 
```

```python
#Resizing Width with preserving aspect ratio
reSizedHeight = 200.0 * image.shape[0] / image.shape[1]
dimNewImage = (200, int(reSizedHeight))
# perform the actual resizing of the image
resizedImage = cv2.resize(image, dimNewImage, interpolation=cv2.INTER_AREA)
displayImage(resizedImage,"Reduced width to 200")
```

```python
#Resizing Height without preserving aspect ratio
dimNewImage = (image.shape[0],100)
# perform the actual resizing of the image
resizedImage = cv2.resize(image, dimNewImage, interpolation=cv2.INTER_AREA)
displayImage(resizedImage,"Reduced height to 100")
```

```python
#Resizing height with preserving aspect ratio
reSizedWidth = 150.0 * image.shape[1] / image.shape[0]
dimNewImage = (int(reSizedWidth),150)
# perform the actual resizing of the image
resizedImage = cv2.resize(image, dimNewImage, interpolation=cv2.INTER_AREA)
displayImage(resizedImage,"Reduced height to 150")
```

### Interpolation

```python
image = cv2.imread("./data/nature_image.jpg")
displayImage(image, "original")
```

```python
# construct the list of interpolation methods
methods = [
    ("cv2.INTER_NEAREST", cv2.INTER_NEAREST),
    ("cv2.INTER_LINEAR", cv2.INTER_LINEAR),
    ("cv2.INTER_AREA", cv2.INTER_AREA),
    ("cv2.INTER_CUBIC", cv2.INTER_CUBIC),
    ("cv2.INTER_LANCZOS4", cv2.INTER_LANCZOS4)]
 
# loop over the interpolation methods
for (name, method) in methods:
    # increase the size of the image by 3x using the current interpolation
    # method
    print("name: {} method:{}".format(name, method))
    reSizedHeight = 400.0 * image.shape[0] / image.shape[1]
    dimNewImage = (400, int(reSizedHeight))
    # perform the actual resizing of the image
    resizedImage = cv2.resize(image, dimNewImage, interpolation=method)
    displayImage(resizedImage,"Method: {}".format(name))
   
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayImage(image, title):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
#load image from disk
image = cv2.imread("./data/person_image.jpg")
displayImage(image, "original")



```

```python
#define translation matrix
# M = [1 0 tx
#.     0 1 ty ]
M = np.float32([[1, 0, -125], [0, 1, -250]])
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
displayImage(shifted, "shifted")
```
